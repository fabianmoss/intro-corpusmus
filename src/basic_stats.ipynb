# Basic stats

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# sns.set_context("talk")
plt.style.use("ggplot")
```

```python
df = pd.read_csv("../admin/TN_LV142060107_2020-10-26.csv")
```

```python
df.head()
```

```python
fig, axes = plt.subplots(1,2, figsize=(10,4))

df["Gruppe"].value_counts(normalize=True).plot(kind="bar", ax=axes[0])
axes[0].xaxis.set_tick_params(rotation=0)
axes[0].set_xticks([0,1],["Studium Integrale", "Musikwissenschaft"])

df["Semester im Studium"].value_counts().plot(kind="pie", ax=axes[1], cmap="Set2")

plt.savefig("basic_stats.png", dpi=150)
plt.show()
```
